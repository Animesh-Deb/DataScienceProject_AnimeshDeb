#### 1.4 Are there any outliers in the data? Back up your answer with a suitable plot/technique with the help of detailed comments.

```python
# Creating a function to check outliers, outliers are any value that is more that 1.5 times of IQR

def Outlierdetection(df1):
    if (((df1['75%']-df1['25%'])*1.5 + df1['75%'])<df1['max']) or (df1['min']<(df1['25%']-(df1['75%']-df1['25%'])*1.5)):
        a='Yes'
    else :
        a='No'
    return a
    
```

```python
df1['Outlier Detected']=df1.apply(Outlierdetection, axis=1)
```
